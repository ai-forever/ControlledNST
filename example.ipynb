```python
from PIL import Image
import matplotlib.pyplot as plt

from style_transfer import StyleTransfer
```

```python
st = StyleTransfer()
```

```python
content = Image.open('images/content1.jpg')
style = Image.open('images/style1.jpg')

#single image prediction
result = st.predict(content,style,print_every=100,iters=400)
```

```python
plt.figure(figsize=[12,12])
plt.imshow(result)
```

```python
#yield images during style transfer
gen_images = st.predict_iterator(content,style,print_every=50, yield_every=50,iters=200)
imgs = [img for img in gen_images]
```

```python
fig = plt.figure(figsize=[30,30])
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(imgs[i])
```

```python
####
#you can pass paths to images, PIL images or numpy arrays to predict functions
result2_lr, result2_hr = st.predict_hr('images/content2.jpg','images/style2.jpg',print_every=100,scale_img=0.8,hr_scale=1.05,iters=150,hr_iters=200)
```

```python
plt.figure(figsize=[18,18])

plt.subplot(1,2,1).set_title('Basic style transfer result')
plt.imshow(result2_lr)
plt.subplot(1,2,2).set_title('A bit more quality')
plt.imshow(result2_hr)
```
