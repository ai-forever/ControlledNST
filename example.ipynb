```python
from PIL import Image
import matplotlib.pyplot as plt

from style_transfer import StyleTransfer
```

```python
st = StyleTransfer()
```

```python
content = Image.open('images/tubingen.jpg')
style = Image.open('images/starry-night.jpg')

result = st.predict(content,style,scale_img=0.4,print_every=100)
```

```python
plt.figure(figsize=[10,10])
plt.imshow(result)
```

```python
#yield images during style transfer
gen_images = st.predict_iterator(content,style,scale_img=0.4,print_every=50, yield_every=50,iters=200)
imgs = [img for img in gen_images]
```

```python
fig = plt.figure(figsize=[20,20])
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(imgs[i])
```

```python
####
#you can pass paths to images, PIL images or numpy arrays to predict functions
result2_lr, result2_hr = st.predict_hr('images/tubingen.jpg','images/starry-night.jpg',print_every=100,scale_img=0.4,hr_scale=1.0,iters=200,hr_iters=300)
```

```python
plt.figure(figsize=[18,18])

plt.subplot(1,2,1).set_title('Basic style transfer result')
plt.imshow(result2_lr)
plt.subplot(1,2,2).set_title('A bit more quality')
plt.imshow(result2_hr)
```
